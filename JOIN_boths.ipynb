## Cross-validation , Grid-Search

```python
final1 = pd.read_csv(r"C:FINAL.csv")
final1.head()
```

```python
final1.Current_Status.value_counts()
```

```python
final1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

final1[final1.select_dtypes(include = ["object"]).columns] = final1[final1.select_dtypes(include = ["object"]).columns].apply(le.fit_transform)
```

```python
final1_train , final1_test = train_test_split(final1 , test_size = .3)

train_x = final1_train.iloc[: ,0:-1]
train_y = final1_train.iloc[: ,-1 ]

test_x = final1_test.iloc[: ,0:-1]
test_y = final1_test.iloc[: ,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier( criterion = "entropy " , max_depth = 7 , min_samples_split = 40 , class_weight = "balanced" )

from sklearn.model_selection import GridSearchCV

search_dict = {"criterion" :["gini" , "entropy"], # 2
              "max_depth" : range(3,10),          # 7 
              "min_samples_split" :range(25 , 50) } #24

grid = GridSearchCV(dt , param_grid = search_dict)
```

```python
grid.fit(train_x , train_y)

grid.best_params_

pred_grid = grid.predict(test_x)

tab_grid = confusion_matrix(pred_grid , test_y)
tab_grid
```

```python
print(accuracy_score(test_y , pred_grid) )            ###
print('-------------------','----------------')
print(precision_score(pred_grid , test_y))
```
